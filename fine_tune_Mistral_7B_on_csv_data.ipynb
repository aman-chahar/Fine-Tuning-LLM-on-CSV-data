```python
# Drop irrelevant columns
data.drop(['url', 'address', 'phone', 'reviews_list', 'menu_item', 'listed_in(type)', 'listed_in(city)'], axis=1, inplace=True)
```

```python
# Convert 'rate' to numeric (handle non-numeric values)
def convert_rate(x):
    if pd.notnull(x):
        try:
            return float(x.split('/')[0])  # Extract numeric part of the rate
        except ValueError:
            return None  # Return None for non-convertible values like 'NEW'
    else:
        return None

data['rate'] = data['rate'].apply(convert_rate)

# Filter rows where 'rate' is not null and rate >= 3
data = data[data['rate'].notnull() & (data['rate'] >= 3)]
```

```python
#looking for total null values in rest_type column.
print(data['rest_type'].isnull().sum())
```

```python
# Fill null values in rest_type with "casual dining"
data['rest_type'].fillna('casual dining', inplace=True)
```

```python
# Fill null values in dish_liked with "local food"
# self editted
data['dish_liked'].fillna('local food', inplace=True)
```

```python
# Remove all rows with any null values
data = data.dropna()
```

```python
origin_model_path = "mistralai/Mistral-7B-Instruct-v0.1"
model_path = "filipealmeida/Mistral-7B-Instruct-v0.1-sharded"
bnb_config = BitsAndBytesConfig \
              (
                load_in_4bit=True,
                bnb_4bit_use_double_quant=True,
                bnb_4bit_quant_type="nf4",
                bnb_4bit_compute_dtype=torch.bfloat16,
              )
model = AutoModelForCausalLM.from_pretrained (model_path, trust_remote_code=True,
                                              quantization_config=bnb_config,
                                              device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(origin_model_path)
```

```python
text_generation_pipeline = transformers.pipeline(
    model=model,
    tokenizer=tokenizer,
    task="text-generation",
    eos_token_id=tokenizer.eos_token_id,
    pad_token_id=tokenizer.eos_token_id,
    repetition_penalty=1.1,
    return_full_text=False,
    max_new_tokens=300,
    temperature = 0.3,
    do_sample=True,
)
mistral_llm = HuggingFacePipeline(pipeline=text_generation_pipeline)
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
# Load the CSV file
loader = CSVLoader(file_path='/content/file.csv')
data = loader.load()

# Split the documents into smaller chunks
separator=""
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
chunked_docs  = text_splitter.split_documents(data)

# Using HuggingFace embeddings
embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.vectorstores import FAISS
db = FAISS.from_documents(chunked_docs,
                          HuggingFaceEmbeddings(model_name='sentence-transformers/all-mpnet-base-v2'))


# Connect query to FAISS index using a retriever
retriever = db.as_retriever(
    search_type="similarity",
    search_kwargs={'k': 4}
)
```

```python
# Create the Conversational Retrieval Chain
qa_chain = ConversationalRetrievalChain.from_llm(mistral_llm, retriever,return_source_documents=True)
```

```python
#We will run an infinite loop to ask questions to LLM and retrieve answers untill the user wants to quit
import sys
chat_history = []
query = input('Prompt: ')
result = qa_chain({'question': query, 'chat_history': chat_history})
print('Answer: ' + result['answer'] + '\n')
chat_history.append((query, result['answer']))
```
